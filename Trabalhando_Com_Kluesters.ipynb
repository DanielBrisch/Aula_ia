```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

x = np.random.uniform(0, 1, 200)
s = np.random.uniform(0, 1, 200)
print(x)
print(s)
matriz = np.dstack((s, x))
print(matriz)
```

```python
resultado = []
for i in range(1, 11):
  kmeans = KMeans(n_clusters = i, init='k-means++', max_iter = 100)
  kmeans.fit(matriz[0])
  resultado.append(kmeans.inertia_)
print(resultado)
```

```python


plt.plot(range(1, 11), resultado)
plt.title('Metodo de ELbow')
plt.xlabel('clusters')
plt.ylabel('Valores')
plt.show()
```
